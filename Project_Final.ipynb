EXPLORATORY DATA ANALYSIS (EDA)
</h1>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
file_path = "Districtwise-Crime-Against-Children.csv"
df = pd.read_csv(file_path)

# Standardize column names
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")

# Create total_crimes column
crime_columns = [
    'murder',                        
    'kidnapping',                    
    'exposure_and_abandonment', 
    'immoral_traffic_prevention',
    'procuration',
    'pocso',                              
    'attempt_to_commit_murder',                
    'unnatural_offences',            
    'human_trafficking',                 
    'other_crimes',
]

df['total_crimes'] = df[crime_columns].sum(axis=1)
```

1️⃣ UNIVARIATE ANALYSIS

```python
print("Statistical Summary:\n")
print(df[crime_columns + ['total_crimes']].describe())
```

Note: Descriptive statistics (count, mean, standard deviation, minimum, maximum, and quartiles) were  generated for all crime categories and the total_crimes column, helping understand the overall distribution, variability, and central tendency of crime data.

```python
plt.figure()
df['total_crimes'].hist(bins=30)
plt.title("Distribution of Total Crimes")
plt.xlabel("Total Crimes")
plt.ylabel("Frequency")
plt.show()
```

Creates a histogram to visualize the distribution of total crimes across districts, helping identify patterns such as skewness, concentration of values, and possible outliers.

```python
top_districts = df.sort_values(by='total_crimes', ascending=False).head(10)

print("Top 10 Districts by Total Crimes:\n")
print(top_districts[['state_name', 'district_name', 'total_crimes']])
```

Note:Sorts the dataset by total_crimes in descending order and displays the top 10 districts, helping identify the areas with the highest reported crime levels.

2️⃣ BIVARIATE ANALYSIS

```python
state_summary = df.groupby('state_name')['total_crimes'].sum().sort_values(ascending=False)

print("State-wise Total Crimes:\n")
print(state_summary.head(10))
```

Note: Calculated the total crimes for each state and displayed top 10 states, helping identify which states report the highest overall crime level. 

```python
pivot_table = pd.pivot_table(
    df,
    values=crime_columns,
    index='state_name',
    aggfunc='sum'
)

print("Pivot Table (State vs Crime Type):\n")
print(pivot_table.head())
```

Note: Created a pivot table that shows the total number of each crime type for every state, making it easy to compare different crime categories across states.

3️⃣ MULTIVARIATE ANALYSIS

```python
correlation_matrix = df[crime_columns + ['total_crimes']].corr()

print("Correlation Matrix:\n")
print(correlation_matrix)
```

Note: Calculated the correlation between all crime categories and total crimes, helping us understand how strongly different crime types are related to each other.

```python
state_crime_profile = df.groupby('state_name')[crime_columns].sum()

state_crime_percentage = state_crime_profile.div(
    state_crime_profile.sum(axis=1), axis=0
) * 100

print("Crime Percentage Contribution by State:\n")
print(state_crime_percentage.head(5))
```

Note: Calculated the percentage contribution of each crime type within every state, helping us understand which types of crimes are more dominant in each state.

VISUALISATIONS</h1>

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# Define crime columns
crime_columns = [
    'murder',                        
    'kidnapping',                    
    'exposure_and_abandonment', 
    'immoral_traffic_prevention',
    'pocso',                        
    'attempt_to_commit_murder',                
    'unnatural_offences',            
    'human_trafficking',                 
    'other_crimes',
]

# Create total crimes column
df['total_crimes'] = df[crime_columns].sum(axis=1)
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['total_crimes'], bins=30, kde=True)
plt.title("Distribution of Total Crimes per District")
plt.xlabel("Total Crimes")
plt.ylabel("Number of Districts")
plt.show()
```

Insight for the Visual:

<p style="text-align: justify;">The histogram clearly shows a highly right-skewed distribution of total crimes across districts. Most districts report relatively low numbers of total crimes, as indicated by the tall bars concentrated on the left side of the graph. However, there is a long tail extending toward the right, representing a small number of districts with exceptionally high crime counts.

<p style="text-align: justify;">This pattern indicates that crime against children is not evenly distributed, but instead concentrated in a few high-crime districts that significantly influence overall statistics. These districts act as outliers and may require special attention and targeted policy intervention.

```python
plt.figure(figsize=(8,4))
sns.boxplot(x=df['total_crimes'],color = "purple")
plt.title("Box Plot of Total Crimes")
plt.xlabel("Total Crimes")
plt.show()
```

Insight for the Visual:

<p style="text-align: justify;">The box plot clearly indicates that the distribution of total crimes is highly skewed with numerous outliers on the higher end. The majority of districts fall within a relatively low crime range, as shown by the compact box and short lower whisker. However, there are many individual points plotted far beyond the upper whisker, representing districts with exceptionally high crime counts.

<p style="text-align: justify;">This confirms that a small number of districts significantly exceed the typical crime levels and act as extreme outliers. These high-crime districts heavily influence the overall distribution and highlight the need for targeted intervention and focused policy measures in specific regions rather than uniform action across all districts.


```python
plt.figure(figsize=(8,5))
sns.violinplot(x=df['total_crimes'],color="indigo")
plt.title("Violin Plot of Total Crimes per District")
plt.xlabel("Total Crimes")
plt.show()
```

Insight for the Visual:

<p style="text-align: justify;">The violin plot provides a detailed view of the distribution and density of total crimes across districts. The wider portion of the violin near the lower values indicates that most districts report relatively low crime numbers. As the values increase, the violin becomes very narrow, showing that fewer districts fall into the high-crime category.

<p style="text-align: justify;">The long thin tail extending to the right confirms the presence of extreme high-crime districts, reinforcing the right-skewed nature of the distribution. Overall, this visualization clearly demonstrates that crimes against children are concentrated in a small number of districts, while the majority experience comparatively lower crime levels.


```python
state_total = df.groupby('state_name')['total_crimes'].sum().sort_values(ascending=False)

top10_states = state_total.head(10)

plt.figure(figsize=(8,8))
plt.pie(
    top10_states.values,
    labels=top10_states.index,
    autopct='%1.1f%%',
    startangle=140
)

plt.title("Top 10 States by Total Crimes")
plt.show()
```

Insight for the Visual:

<p style="text-align: justify;">The pie chart illustrates the percentage contribution of the top 10 states to the overall total crimes. It clearly shows that Uttar Pradesh contributes the highest share (around 20.5%), followed by Maharashtra (18.1%) and Madhya Pradesh (17.6%). Together, these three states account for more than half of the total crimes among the top 10 states.

<p style="text-align: justify;">States like Delhi and West Bengal also contribute a noticeable portion, while Gujarat, Rajasthan, Bihar, Karnataka, and Chhattisgarh contribute comparatively smaller shares.

<p style="text-align: justify;">Overall, the chart highlights that crime against children is heavily concentrated in a few leading states, emphasizing the need for targeted policy measures and stronger child protection efforts in those high-contributing regions.

```python
plt.figure(figsize=(10,6))
sns.barplot(x=state_total.tail(10).values,
            y=state_total.tail(10).index,color="indigo")
plt.title("Bottom 10 States by Total Crimes")
plt.xlabel("Total Crimes")
plt.ylabel("State")
plt.show()
```

Insight for the Visual:

<p style="text-align: justify;">This bar chart represents the bottom 10 states/UTs* with the lowest total crimes against children.

<p style="text-align: justify;">Among them, Chandigarh reports the highest value within this lower group, followed by Mizoram and Manipur. On the other hand, Lakshadweep records the lowest total crimes, indicating minimal reported cases compared to other regions.

<p style="text-align: justify;">Overall, these states contribute a relatively small share to the national total. However, lower crime numbers may be influenced by factors such as smaller population size, geographic location, administrative structure, or reporting practices. Therefore, while the figures are comparatively low, continuous monitoring and preventive measures remain essential to ensure child safety across all regions.

```python
plt.figure(figsize=(12,8))
corr_matrix = df[crime_columns + ['total_crimes']].corr()
sns.heatmap(corr_matrix,
            annot=True,
            cmap="coolwarm")
plt.title("Correlation Heatmap of Crime Categories")
plt.show()
```

<p style="text-align: justify;">Moderate correlations can be observed between murder, unnatural offences, and total crimes, indicating that violent crime categories tend to rise together. In contrast, **Human Trafficking and Procuration show weak correlations** with most other crimes, suggesting they may occur more independently or at lower frequencies.

<p style="text-align: justify;">Overall, the heatmap confirms that certain crime categories move together and significantly impact total crime levels, highlighting the importance of focusing on highly correlated and high-impact crime types when designing prevention and policy strategies.

These high-crime districts act as statistical outliers and significantly influence the overall averages. This uneven pattern suggests that crimes against children are not uniformly spread but are instead concentrated in specific geographic hotspots, requiring focused and region-specific intervention strategies.

<p style="text-align: justify;">2. At the district level, the identification of the top high-crime districts demonstrates that a limited number of areas contribute disproportionately to the total reported cases. This concentration indicates structural or socio-economic challenges in certain districts that may be contributing to higher crime levels. Factors such as urban density, migration, economic disparity, reporting mechanisms, and law enforcement capacity may influence these patterns. The presence of such crime hotspots highlights the importance of targeted policing, stronger child protection systems, and community-level awareness initiatives rather than relying solely on broad, nationwide policy measures.

<p style="text-align: justify;">3. State-wise aggregation further strengthens this observation, as certain states consistently report significantly higher overall crime totals compared to others. The top-ranking states contribute a large share of total crimes, while several states report comparatively lower figures. This disparity may reflect differences in population size, administrative efficiency, awareness and reporting practices, socio-economic conditions, or effectiveness of preventive measures. Therefore, inter-state comparisons should consider contextual factors, and policy frameworks should be tailored to state-specific conditions rather than adopting a one-size-fits-all approach.

<p style="text-align: justify;">4. An in-depth examination of crime composition across states reveals that the dominant types of crimes vary by region. In several high-crime states, kidnapping and POCSO-related offences form a substantial portion of the total crime profile, while human trafficking appears more concentrated in certain states. This variation indicates that each state possesses a distinct crime structure. Consequently, prevention and enforcement strategies must be customized according to the prevailing crime patterns within each state. For example, states with high kidnapping rates may require enhanced surveillance systems and rapid response mechanisms, whereas states with higher trafficking cases may need stronger border monitoring and organized crime intervention.

<p style="text-align: justify;">5. The correlation analysis provides additional insight into the relationships between different crime categories. A positive association was observed between crimes such as murder and attempt to commit murder, as well as between kidnapping and POCSO cases. These relationships suggest that certain crime types may be interconnected or influenced by similar underlying social, economic, or environmental factors. The presence of these correlations indicates that crime prevention efforts should adopt an integrated approach, addressing root causes rather than treating each crime category in isolation.

<p style="text-align: justify;">6. Outlier analysis through boxplots confirms the presence of extreme districts with exceptionally high crime counts. These districts heavily impact statistical measures such as mean and variance and represent high-risk zones. The identification of such outliers is crucial for policy prioritization, as concentrated intervention in these districts could significantly reduce overall crime levels. Continuous monitoring, enhanced law enforcement allocation, and targeted welfare programs may be required to mitigate risks in these areas.

<p style="text-align: justify;">Overall, the analysis demonstrates that crimes against children exhibit strong regional concentration, variability across states, and meaningful interrelationships between crime categories. The findings emphasize the necessity for data-driven, targeted, and region-specific policy frameworks. By leveraging analytical insights, authorities can allocate resources more efficiently, design focused intervention strategies, and ultimately strengthen child protection mechanisms across high-risk regions.
